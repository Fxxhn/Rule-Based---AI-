```python
import pandas as pd
import os

dataset_path = ''
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        if filename.endswith('.csv'):
            dataset_path = os.path.join(dirname, filename)

columns = ['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety', 'class']
df = pd.read_csv(dataset_path, header=None, names=columns)

def budget_filter(row):
    price_condition = row['buying'] in ['low', 'med']
    maint_condition = row['maint'] in ['low', 'med']
    
    if price_condition and maint_condition:
        return 'acc'
    return 'unacc'

df['Prediction'] = df.apply(budget_filter, axis=1)

correct = len(df[df['Prediction'] == df['class']])
total = len(df)
accuracy = (correct / total) * 100

tp = len(df[(df['Prediction'] == 'acc') & (df['class'].isin(['acc', 'good', 'vgood']))])
fp = len(df[(df['Prediction'] == 'acc') & (df['class'] == 'unacc')])
tn = len(df[(df['Prediction'] == 'unacc') & (df['class'] == 'unacc')])
fn = len(df[(df['Prediction'] == 'unacc') & (df['class'].isin(['acc', 'good', 'vgood']))])

print("\n" + "="*50)
print(f"{'ALGORITHM 1: BUDGET FILTER REPORT':^50}")
print("="*50)
print(f"{'Metric':<25} | {'Total Count':<20}")
print("-" * 50)
print(f"{'Total Samples':<25} | {total}")
print(f"{'Accuracy':<25} | {accuracy:.2f}%")
print("-" * 50)
print(f"{'True Positives (Good Decision)':<25}         | {tp}")
print(f"{'False Positives (Risky Decision)':<25}       | {fp}")
print(f"{'True Negatives (Correct Decision)':<25}      | {tn}")
print(f"{'False Negatives (Missed Opportunity)':<25}   | {fn}")
print("="*50 + "\n")
```
