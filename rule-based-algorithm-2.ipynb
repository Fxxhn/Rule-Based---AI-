```python
import pandas as pd

df = pd.read_csv("/kaggle/input/sms-spam-collection-dataset/spam.csv", encoding='latin-1')
df = df[['v1', 'v2']]
df.columns = ['label', 'message']

def advanced_check_spam(msg):
    msg_lower = msg.lower()
    if any(word in msg_lower for word in ['win', 'free', 'claim', 'urgent', 'prize', 'reward', 'congrats', 'welcome']):
        return 'scam message'
    elif len(msg) > 120:
        return 'scam message'
    elif any(char.isdigit() for char in msg):
        return 'scam message'
    elif 'http' in msg_lower or 'www' in msg_lower:
        return 'scam message'
    else:
        return 'normal message'

df['predicted_label'] = df['message'].apply(advanced_check_spam)
df.head(10)
```
