```python
import pandas as pd
import os

# --- 1. DATA LOADING (Required to define df) ---
dataset_path = ''
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        if filename.endswith('.csv'):
            dataset_path = os.path.join(dirname, filename)

columns = ['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety', 'class']
df = pd.read_csv(dataset_path, header=None, names=columns)
total = len(df) # Define total samples

# --- 2. ALGORITHM 2 LOGIC ---
def family_filter(row):
    price_condition = row['buying'] in ['low', 'med']
    maint_condition = row['maint'] in ['low', 'med']
    size_condition = row['persons'] in ['4', 'more']
    
    if price_condition and maint_condition and size_condition:
        return 'acc'
    return 'unacc'

df['Prediction_2'] = df.apply(family_filter, axis=1)

# --- 3. METRICS CALCULATION ---
correct_2 = len(df[df['Prediction_2'] == df['class']])
accuracy_2 = (correct_2 / total) * 100

tp_2 = len(df[(df['Prediction_2'] == 'acc') & (df['class'].isin(['acc', 'good', 'vgood']))])
fp_2 = len(df[(df['Prediction_2'] == 'acc') & (df['class'] == 'unacc')])
tn_2 = len(df[(df['Prediction_2'] == 'unacc') & (df['class'] == 'unacc')])
fn_2 = len(df[(df['Prediction_2'] == 'unacc') & (df['class'].isin(['acc', 'good', 'vgood']))])

# --- 4. PROFESSIONAL REPORT OUTPUT ---
print("\n" + "="*50)
print(f"{'ALGORITHM 2: FAMILY FILTER REPORT':^50}")
print("="*50)
print(f"{'Metric':<25}       | {'Total':<20}")
print("-" * 50)
print(f"{'Total Samples':<25}       | {total}")
print(f"{'Accuracy':<25}       | {accuracy_2:.2f}%")
print("-" * 50)
print(f"{'True Positives (Good Decision)':<25}        | {tp_2}")
print(f"{'False Positives (Risky Decision)':<25}      | {fp_2}")
print(f"{'True Negatives (Correct One)':<25}          | {tn_2}")
print(f"{'False Negatives (Missed opportunity)':<25}  | {fn_2}")
print("="*50 + "\n")
```
